```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Iris_species.csv')
```

```python
print(dataset)
```

```python
dataset.head()
```

# continuous plot

### displot

```python
sns.displot(data=dataset)
```

```python
sns.displot(dataset['PetalLengthCm'])
```

```python
sns.displot(dataset['PetalWidthCm'],kind='kde')
```

```python
sns.displot(dataset['PetalWidthCm'],kind='ecdf')
```

```python
sns.displot(dataset['PetalWidthCm'],kde=True)
```

### joinplot

```python
sns.jointplot(data=dataset , x= 'PetalWidthCm' , y = 'PetalLengthCm')
```

```python
sns.jointplot(data=dataset , x= 'PetalWidthCm' , y = 'PetalLengthCm',kind= 'reg')
```

```python
sns.jointplot(data=dataset , x= 'PetalWidthCm' , y = 'PetalLengthCm' , color= 'lightgreen')
```

## pairplot

```python
sns.pairplot(data=dataset)
```

```python
sns.pairplot(dataset, hue= 'Id')
```

```python
sns.pairplot(dataset, hue='SepalWidthCm')
```
